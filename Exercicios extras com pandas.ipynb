Crie uma coluna chamada "classificação_valence", categorizando valence em:
"Triste" se < 0.3
"Neutro" se entre 0.3 e 0.6
"Feliz" se > 0.6

```python
dados_musicais["valence"]
```

```python
dados_musicais["valence classification"] = pd.cut(dados_musicais["valence"],bins=[0,0.3,0.6,1],labels=["triste","neutro","feliz"])
dados_musicais["valence classification"]
```
